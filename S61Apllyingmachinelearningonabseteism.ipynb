The best thing in regression is that the model itself will give us a fair indication of which variable are important for the analysis


 Create the targets  (logistic regression is a type of classification so we will be basicly classifying people into clases)

 We will use a methodology and we will make 2 clases (moderately absent,Excessively absent)

 We will take the median b=value of the absenteism time in our cell



 Evrery thing below the median will be norma



```python
# Find the median

data_preprocessed ['Absenteeism Time in Hours'].median()

# Therefore we can say that every one has been absent for more than 3 hours will e=be thought of as excessively absent 

# And if it is less then 3 hours we will assign the value of 0 for it and 1 for more then 3 hours

```

```python
# Create a new variable called target which will measure if a person has been absent for more then 3 hours


# IF it is true return the function =1 otherwise return =0

targets=np.where(data_preprocessed['Absenteeism Time in Hours']> 3, 1, 0)

targets
```

```python
# Adds the targets to the data frame data_preprocessed

data_preprocessed['Excessive Absenteeism'] = targets

data_preprocessed.head()


# Now we have created our target variable and we have added it to our dataframe.
```

Using a median as a cut-off line is numericaly stable and rigid , because by using the median we have simply balanced the data set(half of the targets are zeroes while the other half is 1)

This will prevent our model from learning to output onlu=y 0's and 1's

```python
# Define the number =1 by the total number of target

targets.sum() / targets.shape[0]


# around 46% of the targets are 1 and 56% are zeroes
```

 Remember while balancing the data set the 2 clases need to represent 50% each one, ususaly 60%/40% will work well for logistic regresiion but it is not true for other algorithm such as neural network.(from his experienxce a balance 55% /45% is almost always sufficient)
