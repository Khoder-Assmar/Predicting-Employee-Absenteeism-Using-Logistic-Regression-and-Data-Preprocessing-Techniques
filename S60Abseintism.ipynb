```python
# to apply a certain type of modification iteratively on each value from a series or a column

# in a dataframe, it is a great idea to create a function that can execute this operation 

# for one element and the implement it to all values from the column of interest

# will define a function ,that will directly return the weekday from the data contau=ined in the data column

def date_to_weekday(date_value):
    return date_value.weekday() # Returns the day of the week as an integer (Monday=0, Sunday=6)
```

```python
# The second part regard the creation of the day of the week column 
df_reason_mod['Day of the week'] = df_reason_mod['Date'].apply( date_to_weekday)
```
